```python
# Remplacer les valeurs manquantes par 0 dans les colonnes spécifiées
colonnes_a_remplacer = ["Blessés Légers", "Blessés hospitalisés", "Tué"]
data[colonnes_a_remplacer] = data[colonnes_a_remplacer].fillna(0)
```

```python
# Compter les occurrences de chaque tranche d'âge
tranche_age_counts = data["Tranche d'age"].value_counts()

# Créer un diagramme à barres pour visualiser les modalités de la colonne "Tranche d'âge"
plt.figure(figsize=(10, 6))
tranche_age_counts.plot(kind='bar', color='skyblue')
plt.title('Répartition des données par tranche d\'âge')
plt.xlabel('Tranche d\'âge')
plt.ylabel('Nombre d\'occurrences')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculer le mode de la colonne "Tranche d'âge"
mode_tranche_age = data["Tranche d'age"].mode()[0]
# Remplacer les valeurs nulles dans la colonne "Tranche d'âge" par le mode
data["Tranche d'age"].fillna(mode_tranche_age, inplace=True)
```

```python
# Remplacer les valeurs manquantes dans la colonne "Résumé" par "Données manquantes"
data["Résumé"].fillna("Données manquantes", inplace=True)

# Remplacer les valeurs manquantes dans la colonne "Adresse" par "Adresse inconnue"
data["Adresse"].fillna("Adresse inconnue", inplace=True)
```

```python
# Affichage de la matrice des valeurs manquantes
msno.matrix(data)
plt.show()
```

```python
data = data.drop(labels=["arrondgeo", "Arrondissement.1"], axis=1)
```

```python

# Convertir la colonne 'Date' en type datetime
data['Date'] = pd.to_datetime(data['Date'])

# Convertir les colonnes 'Age', 'Blessés Légers', 'Blessés hospitalisés' et 'Tué' en type entier
colonnes_entier = ['Age', 'Blessés Légers', 'Blessés hospitalisés', 'Tué']
data[colonnes_entier] = data[colonnes_entier].astype(int)
# Convertir les colonnes 'Mode', 'Catégorie', 'Gravité', 'Genre', 'Milieu', 'Tranche d'âge', 'PIM/BD PERIPHERIQUE', 'Résumé', 'Coordonnées', 'Nom arrondissement' et 'arrondgeo' en type str
colonnes_categorie = ['Mode', 'Catégorie', 'Gravité', 'Genre', 'Milieu', 'Tranche d\'age', 'PIM/BD PERIPHERIQUE', 'Résumé', 'Coordonnées', 'Nom arrondissement']
# Convertir les colonnes en chaînes de caractères en utilisant la méthode applymap
data[colonnes_categorie] = data[colonnes_categorie].applymap(str)
```

```python
import pandas as pd 
import missingno as msno
import matplotlib.pyplot as plt
import seaborn as sns
import folium

# Grouper les données par arrondissement et compter le nombre d'observations dans chaque arrondissement
observations_par_arrondissement = data.groupby('Nom arrondissement').size()

# Créer une carte Folium centrée sur Paris par exemple
carte = folium.Map(location=[48.8566, 2.3522], zoom_start=12)

# Ajouter des marqueurs pour chaque arrondissement avec le nombre d'observations comme valeur de l'indicateur
for arrondissement, observations in observations_par_arrondissement.items():
    # Obtenir les données pour l'arrondissement en cours
    arrondissement_data = data[data['Nom arrondissement'] == arrondissement]
    
    # Calculer le nombre total d'observations pour l'arrondissement en cours
    total_observations = len(arrondissement_data)
    
    # Créer le contenu de la popup avec les informations souhaitées
    popup_content = f"<div style='max-height: 300px; overflow-y: auto;'><b>Arrondissement {arrondissement}</b><br>"
    popup_content += f"Nombre total d'observations : {total_observations}<br>"
    
    # Ajouter les informations des autres colonnes
    for column in arrondissement_data.columns:
        # Exclure la colonne 'Arrondissement' car elle est déjà incluse dans le titre
        if column != 'Nom arrondissement':
            popup_content += f"{column} : {arrondissement_data[column].iloc[0]}<br>"
    
    popup_content += "</div>"  # Fermeture de la div pour la zone de défilement
    
    # Créer le marqueur cercle avec la popup appropriée
    folium.CircleMarker(
        location=[arrondissement_data['Latitude'].mean(), arrondissement_data['Longitude'].mean()],
        radius=observations/100,  # ajuster la taille des marqueurs en fonction du nombre d'observations
        popup=folium.Popup(popup_content, max_width=300, max_height=300),  # Limiter la taille de la popup
        color='blue',
        fill=True,
        fill_color='blue'
    ).add_to(carte)

carte
```

```python
import streamlit as st

# Convertir la colonne "Date" en format de date
data['Date'] = pd.to_datetime(data['Date'])

# Extraire l'année à partir de la colonne "Date"
data['Année'] = data['Date'].dt.year

# Grouper les données par année et compter le nombre total de victimes pour chaque année
victimes_par_année = data.groupby('Année').agg({'Tué': 'sum', 'Blessés Légers': 'sum', 'Blessés hospitalisés': 'sum'}).reset_index()

# Créer le graphique
fig, ax = plt.subplots(figsize=(10, 6))

# Ajouter les barres pour chaque type de victime
ax.bar(victimes_par_année['Année'], victimes_par_année['Tué'], label='Tué', color='red')
ax.bar(victimes_par_année['Année'], victimes_par_année['Blessés Légers'], bottom=victimes_par_année['Tué'], label='Blessés Légers', color='orange')
ax.bar(victimes_par_année['Année'], victimes_par_année['Blessés hospitalisés'], bottom=victimes_par_année['Tué'] + victimes_par_année['Blessés Légers'], label='Blessés hospitalisés', color='blue')

# Ajouter des étiquettes, des titres et une légende
ax.set_xlabel('Année')
ax.set_ylabel('Nombre de victimes')
ax.set_title('Nombre de victimes par année')
ax.legend()

# Afficher le graphique
st.pyplot(fig)
```
